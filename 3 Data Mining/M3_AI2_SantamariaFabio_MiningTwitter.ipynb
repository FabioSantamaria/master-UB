# 3. Lista de los nombres de los usuarios a los que @rosalia está siguiendo

Para obtener la lista de los nombres de los usuarios a los que @rosalia está siguiendo usamos la función friends del API de *tweepy*. Esta función está limitada a devolver solo 5000 resultados. Para obtener siempre la lista completa debemos usar un Cursor, tal y como se muestra a continuación:

```python
friends_list = []
for page in tweepy.Cursor(api.friends, screen_name=USER_NAME).pages():
    friends_list.extend(page)
    if len(page)>=5000:
        time.sleep(60)
    
for friends in friends_list:
    print(friends.screen_name)
```

Es este caso, vemos que Rosalía sigue a muy poca gente, aproximadamente 10 cuentas. Por lo tanto, no habría sido necesario haber empleado la función Cursor. Una implementación más sencilla sería la siguiente:

```python
for friends in api.friends(user.id):
    print(friends.screen_name)
```

De todas formas, en el caso general, habría que haberlo implementarlo tal y como lo se ha hecho al principio si queremos obtener la lista completa.

#4. ¿Cuál de ellos tiene más seguidores? 

Para contestar esta pregunta nos valemos de la información obtenida en la consulta anterior. Recorremos la lista asignando en la variable *friend_with_more_followers* aquel usuario que tenga más seguidores.

```python
max = 0
friend_with_more_followers = ""
for friend in friends_list:
    num_friends = friend.followers_count
    if (num_friends > max):
        max = num_friends
        friend_with_more_followers = friend

print("El usuario seguido por Rosalía que tiene más seguidores es " + friend_with_more_followers.screen_name + " con un total de " + str(max)  + " seguidores.")
```
