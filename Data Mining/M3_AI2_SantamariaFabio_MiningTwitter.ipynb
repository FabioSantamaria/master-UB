# 3. Lista de los nombres de los usuarios a los que @rosalia estÃ¡ siguiendo

Para obtener la lista de los nombres de los usuarios a los que @rosalia estÃ¡ siguiendo usamos la funciÃ³n friends del API de *tweepy*. Esta funciÃ³n estÃ¡ limitada a devolver solo 5000 resultados. Para obtener siempre la lista completa debemos usar un Cursor, tal y como se muestra a continuaciÃ³n:

```python
friends_list = []
for page in tweepy.Cursor(api.friends, screen_name=USER_NAME).pages():
    friends_list.extend(page)
    if len(page)>=5000:
        time.sleep(60)
    
for friends in friends_list:
    print(friends.screen_name)
```

Es este caso, vemos que RosalÃ­a sigue a muy poca gente, aproximadamente 10 cuentas. Por lo tanto, no habrÃ­a sido necesario haber empleado la funciÃ³n Cursor. Una implementaciÃ³n mÃ¡s sencilla serÃ­a la siguiente:

```python
for friends in api.friends(user.id):
    print(friends.screen_name)
```

De todas formas, en el caso general, habrÃ­a que haberlo implementarlo tal y como lo se ha hecho al principio si queremos obtener la lista completa.
